## Grafica Histograma

```python
# librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
## creamos los datos para el histagrama

# numero de datos totales
n = 1000

# generamos los datos - usamos la distribuci√≥n log-normal
data = np.exp( np.random.randn(n)/2 )
plt.plot(data,'o')
```

```python
# mostramos el histograma

# numero de bins del histograma
k = 100

plt.hist(data,bins=k)
plt.show()
```

```python
# otra opcion
y,x = np.histogram(data,bins=k)

# bin centers (no me queda claro porque hace esto V28-16:20)
xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
```
