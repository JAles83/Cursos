# Seccion: Visualizacion de Datos

En esta seccion mostraremos como se implementa el grafico de caja de Bigotes en Python

```python
# librerias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns #modulo que posee graficas muy interesantes
```

```python
## creamos los datos con los que trabajaremos

# tamanio de los datos
m = 30 # rows
n =  6 # columns

# generamos los datos
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2

# Mostramos los datos
print(data)
```

```python
# metodo para mostrar la caja de bigotes
plt.boxplot(data)
plt.show()
```

```python
# hacemos lo mismo pero con el paquete de seaborn
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
# podemos usar un dataframe
df = pd.DataFrame(data,columns=['zero','one','two','three','four','five'])
sns.boxplot(data=df,orient='h')
plt.show()
```

## Distribucion Normal vs Distribucion Uniforme

```python
## creamos los datos con los que trabajaremos

# tamanio de los datos
m = 30 # rows
n =  6 # columns


# generamos los datos
data1 = np.zeros((m,n)) # Distribucion Normal
data2 = np.zeros((m,n)) # Distribucion Uniforme

for i in range(n):
    data1[:,i] = np.random.randn(m)
    data2[:,i] = np.random.rand(m)
```

```python
## Distribucion Normal
sns.boxplot(data=data1,orient='v')
plt.show()
```

```python
## Distribucion Uniforme
sns.boxplot(data=data2,orient='v')
plt.show()
```
