## Grafica de Lineas

```python
# librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
## creamos los datos para la grafica

# numero de puntos
n = 1000

# generamos una distribucion log-normal
data1 = np.exp( np.random.randn(n)/2 )
data2 = np.exp( np.random.randn(n)/10 )
data3 = np.exp( np.random.randn(n)/2 + 1 )
```

```python
## puntos de sus histogramas

# numero de bins del histograma
k = 20

plt.hist(data1,bins=k)
plt.hist(data2,bins=k)
plt.hist(data3,bins=k)

plt.show()
```

Podemos observar que la manera elegida no es la mejor para mostrar los datos, ya que estos aparecen superpuestos. En estos casos, lo mejor es optar por el grafico de lineas

```python
# Creamos el histograma de cada conjunto de datos
y1,x1 = np.histogram(data1,bins=k)
xx1 = (x1[0:-1] + x1[1:]) / 2
y1 = 100*y1 / sum(y1) # conversion a probabilidad

y2,x2 = np.histogram(data2,bins=k)
xx2 = (x2[0:-1] + x2[1:]) / 2
y2 = 100*y2 / sum(y2) # conversion a probabilidad

y3,x3 = np.histogram(data3,bins=k)
xx3 = (x3[0:-1] + x3[1:]) / 2
y3 = 100*y3 / sum(y3) # conversion a probabilidad



# mostramos los graficos
plt.plot(xx1,y1,'s-',label='data1')
plt.plot(xx2,y2,'o-',label='data2')
plt.plot(xx3,y3,'^-',label='data3')

plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```
