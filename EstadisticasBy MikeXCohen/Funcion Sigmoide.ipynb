Es aquella que tiene la forma $$P(t)=\frac{1}{1+e^{-t}}$$

```python
# PAQUETES
import numpy as np
import sympy as sym
import sympy.plotting.plot as symplot
import math
```

```python
# VARIABLES
t = sym.symbols('t') 
```

# Representacion Grafica

```python
y1 = 1/(1+math.e**t)
y2 = 2/(1+math.e**t)
y3 = 1/(1+math.e**t)-1

p = symplot(y1, show=False)
p2 = symplot(y2, show=False)
p3 = symplot(y3, show=False)

p.extend(p2)
p.extend(p3)

p[0].label = '$\\frac{1}{1+e^{-t}}$'
p[1].label = '$\\frac{2}{1+e^{-t}}$'
p[2].label = '$\\frac{1}{1+e^{-t}}-1$'

p.legend=True

p.show()

print(p)
```
