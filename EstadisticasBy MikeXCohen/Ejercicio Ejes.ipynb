En esta libreta representaremos dos funciones en dos graficos diferentes:
* En uno con ejes lineales
* En otro con ejes logaritmicos

```python
# librerias
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# creamos los datos para la grafica
x = np.linspace(0,10,100)
y_1 = x
y_2 = math.e**x
```
